```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
torch.set_float32_matmul_precision('high')  # or 'medium' 

model_id = "./download_llms/medgemma-27b-text-it"

llm = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
import ast
import torch._dynamo
torch._dynamo.config.suppress_errors = True


def format_features_for_prompt(features_str):
    """
    Convert the list of features in string form to natural language form.
    Input: "[('Triage RespiratoryRate', '30.0 rpm'), ('Triage SBP', '217.0 mmHg'), ...]" "
    Output: "Triage Respiratory Rate: 30.0 rpm; Triage Systolic Blood Pressure: 217.0 mmHg;" ..."
    """
    try:
        features_list = ast.literal_eval(features_str)
        return "; ".join([f"{k}: {v}" for k, v in features_list])
    except Exception as e:
        print("Feature analysis failed:", e)
        return ""

def get_decision_reason(response):
    """
    Extract "yes", "no", "uncertain" and their corresponding reasons from the LLM response.
    """
    # response = response.strip().lower()
    for keyword in ['yes:', 'no:', 'uncertain:']:
        if response.startswith(keyword):
            decision = keyword[:-1]
            reason = response[len(keyword):].strip()
            return decision, reason
    return "uncertain", "Unable to parse decision from response."

def evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, Feature_Values, llm, tokenizer):
    """
    Use LLM to determine whether the death risk warning is a false alarm: yes: consider the model's judgment to be correct; no: consider the model's judgment to be incorrect
    """

    prompt_template_en = f"""
    You are a clinical medicine expert responsible for evaluating whether a high-risk in-hospital mortality alert should be issued for Emergency Department inpatients.

    The machine learning prediction model assigned a {risk_level} level of mortality risk alert to an inpatient. 
    This patient has the following clinical characteristics: {Feature_Values}; dangerous features are: {Dangerous_features}; and the protective features are: {Protective_features}

    Based on your clinical knowledge, please assess the model's risk alert level: If you agree with the alert level, respond with "yes"; 
    If you believe the alert is inaccurate, respond with "no"; If you're uncertain, respond with "uncertain".
    Please respond with only one of: yes / no / uncertain, followed by a brief clinical justification.

    Example response format:
    yes: The patient presents with multiple risk factors including hypoalbuminemia, hyperbilirubinemia, and coagulopathy, consistent with high-risk status
    no: The patient is hemodynamically stable, with no abnormal inflammatory or organ dysfunction markers; risk appears overestimated
    uncertain: The patient has conflicting findings (e.g., elevated troponin but normal cardiac function), requiring further evaluation
    """

    # Construct the chat format
    messages = [
        {"role": "system", "content": "You are a clinical assistant specialized in ED mortality triage."},
        {"role": "user", "content": prompt_template_en}
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = tokenizer([text], return_tensors="pt").to(llm.device)

    generated_ids = llm.generate(
        **model_inputs,
        max_new_tokens=200,
        # temperature=0.2,
        temperature=0.1,
    )

    generated_ids = [output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    response = response.strip()
    
    # Extract the decisions and reasons
    decision, reason = get_decision_reason(response)
    return decision, reason
```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

llmName = 'llama3.3:70b'

llm = OllamaLLM(model=llmName, temperature=0.1, max_tokens=200)

def get_decision_reason(response):
    """Extract decisions and reasons from the responses of the LLM."""
    lines = response.strip().split('\n')
    for line in reversed(lines):
        if ':' in line:
            decision, reason = line.split(':', 1)
            return decision.strip().lower(), reason.strip()
    
    return "uncertain", response.strip()

def evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, Feature_Values, llm):
    """
    Using the English prompt template, have the LLM determine whether the current patient is a false alarm.
    """
    ############# yes/no/uncertain #############
    prompt_template_en = f"""
    You are a clinical medicine expert responsible for evaluating whether a high-risk in-hospital mortality alert should be issued for Emergency Department inpatients.

    The machine learning prediction model assigned a {risk_level} level of mortality risk alert to an inpatient. 
    This patient has the following clinical characteristics: {Feature_Values}; dangerous features are: {Dangerous_features}; and the protective features are: {Protective_features}

    Based on your clinical knowledge, please assess the model's risk alert level: If you agree with the alert level, respond with "yes"; 
    If you believe the alert is inaccurate, respond with "no"; If you're uncertain, respond with "uncertain".
    Please respond with only one of: yes / no / uncertain, followed by a brief clinical justification.

    Example response format:
    yes: The patient presents with multiple risk factors including hypoalbuminemia, hyperbilirubinemia, and coagulopathy, consistent with high-risk status
    no: The patient is hemodynamically stable, with no abnormal inflammatory or organ dysfunction markers; risk appears overestimated
    uncertain: The patient has conflicting findings (e.g., elevated troponin but normal cardiac function), requiring further evaluation
    """

    prompt = ChatPromptTemplate.from_template(prompt_template_en)
    chain = prompt | llm
    response = chain.invoke({"risk_level": risk_level, "Dangerous_features": Dangerous_features, "Protective_features":Protective_features, "Feature_Values": Feature_Values})  # 变量传入

    decision, reason = get_decision_reason(response)
    return decision, reason
```

```python
## test
patient_idx = 0

patient_id = high_risk_patients.iloc[patient_idx]['stay_id']
risk_level = high_risk_patients.iloc[patient_idx]['risk_level']
feature_list = high_risk_patients.iloc[patient_idx]['Feature_Values']  
predicted_risk = round(high_risk_patients.iloc[patient_idx]['y_tepred'] * 100,2)  # 模型输出预测风险为0~1，转成100%
Dangerous_features = high_risk_patients.iloc[patient_idx]['Dangerous features']
Protective_features = high_risk_patients.iloc[patient_idx]['Protective features']

decision, reason = evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, feature_list, llm)
print(f"Decision: {decision}")
print(f"Reason: {reason}")
```

```python
import pandas as pd
import os
from tqdm import tqdm  
from concurrent.futures import ThreadPoolExecutor, as_completed

# save path
save_path = cd + 'llm_results_all_high_risk_Ext/llama/' + db + '/'
if not os.path.exists(os.path.dirname(save_path)):
    os.makedirs(os.path.dirname(save_path))

# define a function to process each patient
def process_patient(i):
    patient_row = high_risk_patients.iloc[i]

    patient_id = patient_row['stay_id']
    risk_level = patient_row['risk_level']
    feature_list = patient_row['Feature_Values']
    # predicted_risk = round(patient_row['y_tepred'] * 100, 2)
    y_test = patient_row['y_test']
    Dangerous_features = patient_row['Dangerous features']
    Protective_features = patient_row['Protective features']


    decision, reason = evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, feature_list, llm)

    return {
        'stay_id': patient_id,
        'y_test': y_test,
        'risk_level': risk_level,
        'LLM_decision': decision,
        'LLM_reason': reason
    }

# ===========================
# process all patients: using multi-threading
# ===========================
results = []
with ThreadPoolExecutor(max_workers=6) as executor:  
    futures = [executor.submit(process_patient, i) for i in range(len(high_risk_patients))]
    for f in tqdm(as_completed(futures), total=len(futures)):
        results.append(f.result())

# DataFrame
llm_results_df = pd.DataFrame(results)

# save
llm_results_df.to_csv(save_path + 'llm_false_alarm_analysis.csv', index=False)
```

```python
import os
import pandas as pd
import numpy as np

db = 'SZBA'  # example database name

cd = './LLM_Risk_Stratification/'
file_path = cd + 'LLM_input_data/' + db +'_Ext' + '/' + 'llm_input_data.csv'  

## select high-risk patients
feature_values = pd.read_csv(file_path)
feature_values = feature_values[feature_values['risk_level'] == 'high-risk']  
feature_values = feature_values.reset_index(drop=True) 
feature_values = feature_values[['stay_id','Feature_Values']]
```

```python
high_risk_patients = high_risk_patients.merge(feature_values,on='stay_id',how='inner')
# print(high_risk_patients.head())
```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

llmName = 'qwen3:32b'

llm = OllamaLLM(model=llmName, temperature=0.1, max_tokens=200)

def get_decision_reason(response):
    """Extract decisions and reasons from the responses of the LLM."""
    lines = response.strip().split('\n')
    for line in reversed(lines):
        if ':' in line:
            decision, reason = line.split(':', 1)
            return decision.strip().lower(), reason.strip()
    
    return "uncertain", response.strip()

def evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, Feature_Values, llm):
    """
    prompt
    """
    ############# yes/no/uncertain #############
    prompt_template_en = f"""
    You are a clinical medicine expert responsible for evaluating whether a high-risk in-hospital mortality alert should be issued for Emergency Department inpatients.

    The machine learning prediction model assigned a {risk_level} level of mortality risk alert to an inpatient. 
    This patient has the following clinical characteristics: {Feature_Values}; dangerous features are: {Dangerous_features}; and the protective features are: {Protective_features}

    Based on your clinical knowledge, please assess the model's risk alert level: If you agree with the alert level, respond with "yes"; 
    If you believe the alert is inaccurate, respond with "no"; If you're uncertain, respond with "uncertain".
    Please respond with only one of: yes / no / uncertain, followed by a brief clinical justification.

    Example response format:
    yes: The patient presents with multiple risk factors including hypoalbuminemia, hyperbilirubinemia, and coagulopathy, consistent with high-risk status
    no: The patient is hemodynamically stable, with no abnormal inflammatory or organ dysfunction markers; risk appears overestimated
    uncertain: The patient has conflicting findings (e.g., elevated troponin but normal cardiac function), requiring further evaluation
    """

    prompt = ChatPromptTemplate.from_template(prompt_template_en)
    chain = prompt | llm
    response = chain.invoke({"risk_level": risk_level, "Dangerous_features": Dangerous_features, "Protective_features":Protective_features, "Feature_Values": Feature_Values})  # 变量传入

    decision, reason = get_decision_reason(response)
    return decision, reason
```

```python
## test
patient_idx = 0

patient_id = high_risk_patients.iloc[patient_idx]['stay_id']
risk_level = high_risk_patients.iloc[patient_idx]['risk_level']
feature_list = high_risk_patients.iloc[patient_idx]['Feature_Values']  
predicted_risk = round(high_risk_patients.iloc[patient_idx]['y_tepred'] * 100,2)  
Dangerous_features = high_risk_patients.iloc[patient_idx]['Dangerous features']
Protective_features = high_risk_patients.iloc[patient_idx]['Protective features']

decision, reason = evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, feature_list, llm)
print(f"Decision: {decision}")
print(f"Reason: {reason}")
```

```python
import pandas as pd
import os
from tqdm import tqdm  
from concurrent.futures import ThreadPoolExecutor, as_completed

# save path
save_path = cd + 'llm_results_all_high_risk_Ext/qwen/' + db + '/'
if not os.path.exists(os.path.dirname(save_path)):
    os.makedirs(os.path.dirname(save_path))

# define a function to process each patient
def process_patient(i):
    patient_row = high_risk_patients.iloc[i]

    patient_id = patient_row['stay_id']
    risk_level = patient_row['risk_level']
    feature_list = patient_row['Feature_Values']
    # predicted_risk = round(patient_row['y_tepred'] * 100, 2)
    y_test = patient_row['y_test']
    Dangerous_features = patient_row['Dangerous features']
    Protective_features = patient_row['Protective features']


    decision, reason = evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, feature_list, llm)

    return {
        'stay_id': patient_id,
        'y_test': y_test,
        'risk_level': risk_level,
        'LLM_decision': decision,
        'LLM_reason': reason
    }

# ===========================
# process all patients: using multi-threading
# ===========================
results = []
with ThreadPoolExecutor(max_workers=6) as executor:  # 可以调整线程数，如 4~8
    futures = [executor.submit(process_patient, i) for i in range(len(high_risk_patients))]
    for f in tqdm(as_completed(futures), total=len(futures)):
        results.append(f.result())

# DataFrame
llm_results_df = pd.DataFrame(results)

# save results
llm_results_df.to_csv(save_path + 'llm_false_alarm_analysis.csv', index=False)
```

```python
def get_decision_reason(response):
    """
    Extract decisions and reasons from the responses of the LLM.
    """
    # response = response.strip().lower()
    for keyword in ['yes:', 'no:', 'uncertain:']:
        if response.startswith(keyword):
            decision = keyword[:-1]
            reason = response[len(keyword):].strip()
            return decision, reason
    return "uncertain", "Unable to parse decision from response."

def evaluate_risk_alert_with_llm(risk_level, Dangerous_features, Protective_features, Feature_Values):
    """
    Evaluate whether a high-risk in-hospital mortality alert should be issued for Emergency Department inpatients using the LLM.
    """
    

    prompt_template_en = f"""
    You are a clinical medicine expert responsible for evaluating whether a high-risk in-hospital mortality alert should be issued for Emergency Department inpatients.

    The machine learning prediction model assigned a {risk_level} level of mortality risk alert to an inpatient. 
    This patient has the following clinical characteristics: {Feature_Values}; dangerous features are: {Dangerous_features}; and the protective features are: {Protective_features}

    Based on your clinical knowledge, please assess the model's risk alert level: If you agree with the alert level, respond with "yes"; 
    If you believe the alert is inaccurate, respond with "no"; If you're uncertain, respond with "uncertain".
    Please respond with only one of: yes / no / uncertain, followed by a brief clinical justification.

    Example response format:
    yes: The patient presents with multiple risk factors including hypoalbuminemia, hyperbilirubinemia, and coagulopathy, consistent with high-risk status
    no: The patient is hemodynamically stable, with no abnormal inflammatory or organ dysfunction markers; risk appears overestimated
    uncertain: The patient has conflicting findings (e.g., elevated troponin but normal cardiac function), requiring further evaluation
    """

    # Construct the chat format
    messages_types = [
        {"role": "system", "content": "You are a clinical assistant specialized in ED mortality triage."},
        {"role": "user", "content": prompt_template_en}
    ]  

    response = client.chat.completions.create(
        model="deepseek-v3-0324",  
        messages=messages_types,
        temperature=0.1,  
        max_tokens=200,
        stream=False
    )
    response = response.choices[0].message.content.strip()  

    # extract decisions and reasons
    decision, reason = get_decision_reason(response)
    return decision, reason
```
